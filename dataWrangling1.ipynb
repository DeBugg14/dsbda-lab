```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/DeBugg14/te-sppu-lab/main/Data-Science/Prac1/prac1.csv'
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
for col in ['stroke','horsepower','peak-rpm']:
    df[col].replace(np.nan,df[col].astype('float').mean(axis=0),inplace=True)
df.isnull().sum()
```

```python
df = df.dropna() #dropping the Nan object
```

```python
df.dtypes
```

```python
df.shape
```

```python
df['city-L/100km'] = round(df['city-L/100km'],4) #data formatting
```

```python
df['horsepower'] = df['horsepower']/df['horsepower'].max() #data normalisation
```

```python
df['engine-location'].replace(['front','rear'],[1,0],inplace=True)
df
```
