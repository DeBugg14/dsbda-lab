```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.isnull() #checking missing/inconsistences
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['math score']) #detecting the outliers
```

```python
arr = np.where(df['math score']<20) #storing the outliers index in arr
for i in arr:
    df.drop(i,inplace = True) #dropping the outliers
```

```python
for col in ['math score','reading score','writing score']: #data transformation
    df[col] = df[col]/df[col].max()

df
```
