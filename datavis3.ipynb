```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset('iris')



```

```python
#List down the features and their types (e.g., numeric, nominal) available in the dataset
df.head()
```

```python

#Create a histogram for each feature in the dataset to illustrate the feature distributions. 
figure,axes = plt.subplots(2,2,figsize=(16,9))
sns.histplot(df['sepal_length'],ax=axes[0,0])
sns.histplot(df['sepal_width'],ax=axes[0,1])
sns.histplot(df['petal_length'],ax=axes[1,0])
sns.histplot(df['petal_width'],ax=axes[1,1])
```

```python
figure,axes = plt.subplots(2,2,figsize=(16,9))

sns.boxplot(y='sepal_length',x='species',data=df,ax=axes[0,0])
```
