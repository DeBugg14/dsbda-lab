```python
import pandas as pd
import numpy as np
```

```python
url_link = 'https://raw.githubusercontent.com/DeBugg14/te-sppu-lab/main/Data-Science/Prac1/prac1.csv'
df = pd.read_csv(url_link)
```

```python
df.isnull() #checking missing/insconsistence values
```

```python
df.isnull().sum() #checking total missing/inconsistent values
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0) #filling missind data
df['stroke'].replace(np.nan,avg_stroke,inplace=True)
```

```python
df.isnull().sum()
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes  #checking data types
```

```python
df['stroke'] = df['stroke'].astype('object') #changing datatypes of a dataset
df.dtypes
```

```python
df.info()
```

```python
df['engine-location'].replace(['front','rear'],[0,1],inplace=True)  #changing catergorical values into quantitative values
df
```
